```python
import warnings
warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
%pylab inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## пожалуйста, укажите путь к csv-датафрему здесь
df = pd.read_csv(r"C:\Users\Artem\Desktop\py_proj\latin\lat_df.csv") 
```

```python
## get rhymes type plot
gl = df.head(1)
vals = [int(gl['2x_rhymes'].get(0)), int(gl['3x_rhymes'].get(0)),
       int(gl['4+_rhymes'].get(0)), int(gl['Full_rhymes'].get(0))]

inf = ['2x_rhymes', '3x_rhymes', '4+_rhymes', 'Full_rhymes']

plot = sns.barplot(x=vals, y=inf)
fig = plot.get_figure()

## здесь и везде далее нужно указывать путь, куда сохранять png графика
fig.savefig(r'C:\Users\Artem\Desktop\py_proj\latin\plot_data\rhm_type.png')
```

```python
## get 3x_rhymes plot

id_lst = df['Id'][1:]
val_lst = df['3x_rhymes'][1:]
plot = sns.barplot(x=val_lst, y=id_lst)

fig = plot.get_figure()
fig.savefig(r'C:\Users\Artem\Desktop\py_proj\latin\plot_data\3x_rhms.png')
```

```python
## get Full_rhymes plot

id_lst = df['Id'][1:]
val_lst = df['Full_rhymes'][1:]
plot = sns.barplot(x=val_lst, y=id_lst)

fig = plot.get_figure()
fig.savefig(r'C:\Users\Artem\Desktop\py_proj\latin\plot_data\full_rhms.png')
```

```python
## get Rhymes_ratio plot
def to_float(val):
    return float(val[:-1])

df['Rhyme_ratio_float'] = df['Rhyme_ratio'].apply(to_float)
id_lst = df['Id'][1:]
val_lst = df['Rhyme_ratio_float'][1:]
plot = sns.barplot(x=val_lst, y=id_lst)

fig = plot.get_figure()
fig.savefig(r'C:\Users\Artem\Desktop\py_proj\latin\plot_data\rhms_ratio.png')
```

```python
## get Rhymes_total plot

id_lst = df['Id'][1:]
val_lst = df['Rhymes_total'][1:]
plot = sns.barplot(x=val_lst, y=id_lst)

fig = plot.get_figure()
fig.savefig(r'C:\Users\Artem\Desktop\py_proj\latin\plot_data\rhms_total.png')
```

```python
## intersect Rhymes_ratio and Rhymes_total
def to_float(val):
    return float(val[:-1])

df['Rhyme_ratio_float'] = df['Rhyme_ratio'].apply(to_float)
inf_lst = df['Rhymes_total'][1:]
val_lst = df['Rhyme_ratio_float'][1:]
plot = sns.barplot(x=inf_lst, y=val_lst)

fig = plot.get_figure()
fig.savefig(r'C:\Users\Artem\Desktop\py_proj\latin\plot_data\rat_rhm_intersect.png')
```
